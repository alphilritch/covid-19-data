```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
current_path = os.getcwd()
current_path
```

```python
data = pd.read_csv('us-states.csv')
print(data)
```

```python
df = pd.DataFrame(pd.read_csv("us-states.csv"))
print(df.head())
```

```python
Washington = df.loc[df['state'] == 'Washington']
print(Washington)
fig, ax1 = plt.subplots()
x = Washington['date']
y1 = Washington['cases']
y2 = Washington['deaths']
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
```

```python
Alaska = df.loc[df['state'] == 'Alaska']
print(Alaska)
fig, ax1 = plt.subplots()
x = Alaska['date']
y1 = Alaska['cases']
y2 = Alaska['deaths']
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
```
